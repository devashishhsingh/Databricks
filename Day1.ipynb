```python

data = [("iphone", 999), ("Samsung", 799), ("Pixel", 699)]
df = spark.createDataFrame(data, ["brand", "price"])
df.show()
df.filter(df.price > 700).show()
```

```python
df.printSchema()
```

```python
display(df.where(df["price"] > 791))
```

```python
from pyspark.sql.functions import desc, asc
display(df.select("brand", "price").orderBy(asc("price")))
```

```python
subsetDF = df.filter((df["brand"] == "Pixel") & (df["price"] > 691))
display(subsetDF)
```
