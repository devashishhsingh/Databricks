```python
customers_data = [
    (1, "Amit", "Delhi"),
    (2, "Priya", "Mumbai"),
    (3, "Ravi", "Chennai"),
    (4, "Neha", "Bangalore"),
    (5, "Priyanka", "Delhi"),
    (6, "Rajesh", "Mumbai")
]
customers_columns = ["customer_id", "customer_name", "city"]
df_customers = spark.createDataFrame(customers_data, customers_columns)
df_customers.show()
```

```python
orders_data = [
    (101, 1, "2024-01-01", 2500),
    (102, 1, "2024-01-05", 4000),
    (103, 2, "2024-01-03", 1500),
    (104, 3, "2024-01-10", 6000),
    (105, 4, "2024-01-12", 3500)
]

orders_columns = ["order_id", "customer_id", "order_date", "order_amount"]

df_orders = spark.createDataFrame(orders_data, orders_columns)

df_orders.show()
```

```python
products_data = [
    (1001, "Laptop", "Electronics"),
    (1002, "Mobile", "Electronics"),
    (1003, "Shoes", "Fashion"),
    (1004, "Watch", "Accessories")
]

products_columns = ["product_id", "product_name", "category"]

df_products = spark.createDataFrame(products_data, products_columns)

df_products.show()
```

```python
order_items_data = [
    (101, 1001, 1),
    (101, 1003, 2),
    (102, 1002, 1),
    (103, 1004, 1),
    (104, 1001, 1),
    (105, 1003, 1)
]

order_items_columns = ["order_id", "product_id", "quantity"]

df_order_items = spark.createDataFrame(order_items_data, order_items_columns)

df_order_items.show()
```

```python
df_customers.printSchema()
df_orders.printSchema()
df_products.printSchema()
df_order_items.printSchema()
```

```python
df_inner = df_customers.join(df_orders,
                             df_customers.customer_id == df_orders.customer_id,
                             "inner")
df_inner.show()
```

```python
df_right = df_customers.join(
    df_orders,
    df_customers.customer_id == df_orders.customer_id,
    "right"
)

df_right.show()
```

```python
df_outer = df_customers.join(
    df_orders,
    df_customers.customer_id == df_orders.customer_id,
    "outer"
)

df_outer.show()
```

```python
df_full_orders = df_orders.join(df_order_items, "order_id", "inner").join(df_products, "prodcut_id", "inner")

df_full_orders.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum, rank, row_number, dense_rank


df_orders.show()
```

```python
window_spec = Window.partitionBy("customer_id").orderBy("order_date").rowsBetween(Window.unboundedPreceding, Window.currentRow)
```

```python
df_running_total = df_orders.withColumn(
    "running_total",
    sum("order_amount").over(window_spec)
)

df_running_total.show()
```
