```python
orders_data = [
    (101, 1, "2024-01-01", 2500),
    (102, 1, "2024-01-05", 4000),
    (103, 2, "2024-01-03", 1500)
]

columns = ["order_id", "customer_id", "order_date", "order_amount"]

df_orders = spark.createDataFrame(orders_data, columns)
```

```python
df_orders.write \
    .format("delta") \
    .mode("overwrite") \
    .save("/tmp/delta/orders")
```
